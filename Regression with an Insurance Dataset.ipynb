```python
def numerical_plot(df):
    col_numerical=df.select_dtypes(include="number")
    kde_columns=['Age','Annual Income', 'Health Score', 'Vehicle Age', 'Credit Score']
    
    for col in col_numerical.columns:
        plt.figure(figsize=(13, 3))
        plt.subplot(1,2,1)
        if col in kde_columns:
            sns.histplot(col_numerical[col], kde=True, bins=20)
        else: 
            sns.histplot(col_numerical[col], kde=False, bins=20)
        plt.title(f'{col} Distribution')
        plt.xlabel("col")
        plt.ylabel("frequency")
            
        plt.subplot(1,2,2)
        sns.boxplot(x=col_numerical[col], width=0.4)
        plt.title(f'{col} Distribution')
        plt.xlabel(col)

        plt.title(f"Boxplot for {col}") 
        plt.show()

numerical_plot(df)
plt.tight_layout()
```
